```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
```

```python

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df["Category"] = le.fit_transform(df["Category"])  # spam=1, ham=0
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df["Message"], df["Category"], test_size=0.2, random_state=42)
```

```python
max_words = 5000
max_len = 100
```

```python

tokenizer = Tokenizer(num_words=max_words, oov_token="<OOV>")
tokenizer.fit_on_texts(X_train)
```

```python
X_train_seq = tokenizer.texts_to_sequences(X_train)
X_test_seq = tokenizer.texts_to_sequences(X_test)

X_train_pad = pad_sequences(X_train_seq, maxlen=max_len, padding="post")
X_test_pad = pad_sequences(X_test_seq, maxlen=max_len, padding="post")
```

```python
Model = Sequential([
    Embedding(input_dim=max_words, output_dim=128, input_length=max_len),
    LSTM(64, return_sequences=False),
    Dropout(0.3),
    Dense(32, activation="relu"),
    Dense(1, activation="sigmoid")
])
```

```python
Model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)
```

```python
Model.fit(X_train_pad, y_train, epochs=5, batch_size=32, validation_split=0.2)
```

```python
loss, acc = Model.evaluate(X_test_pad, y_test)
print("ANN Accuracy:", acc)
```
